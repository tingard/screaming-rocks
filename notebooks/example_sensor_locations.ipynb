```
%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
from mpl_toolkits import mplot3d
```

```
import numpy as np
import pandas as pd
```

```
eb2 = pd.read_csv('../eb2_array.csv')
```

```
q=np.linspace(0, 2*np.pi, 100)
z=np.linspace(-70, 70, 100)
Qc, Zc=np.meshgrid(q, z)
Xc = 18*np.cos(Qc)
Yc = 18*np.sin(Qc)
```

```
fig = plt.figure(figsize=(9,9))
ax = plt.axes(projection='3d')
ax.scatter3D(eb2['East'], eb2['North'], -eb2['Down'], c='k', s=25, alpha=1.0)
ax.plot_surface(Xc, Yc, Zc, alpha=0.9)
ax.set_xlim(-80,80)
ax.set_ylim(-80,80)
ax.set_zlim(-80,80)
```
